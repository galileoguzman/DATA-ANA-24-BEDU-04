```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Crear un diccionario con los datos.
data = {
    'Nivel_de_estres': [7, 6, 8, 5, 9, 4, 7, 3, 10, 2],
    'Horas_de_sueño': [5, 6, 4, 7, 3, 8, 5, 9, 2, 10],
}

# Crear un DataFrame con los datos.
df = pd.DataFrame(data)

# Mostrar el DataFrame.
df.head()
```

```python
X = df[['Nivel_de_estres']]
Y = df[['Horas_de_sueño']]

model = LinearRegression()
model.fit(X, Y)
```

```python
coeficiente = model.coef_[0]
intercepto = model.intercept_

print(f'Coef: {coeficiente}')
print(f'Intercept: {intercepto}')
```

# Visualizacion

```python
plt.scatter(
    X, 
    Y,
    color='blue',
    label='Datos'
)

plt.plot(X, model.predict(X), color='red', linewidth=2, label='REgresion')
plt.title('Regresion lineal sencilla')
plt.xlabel('Nivel de estres')
plt.ylabel('Horas de sueno')
plt.legend()
plt.grid(True)
plt.show()
```
